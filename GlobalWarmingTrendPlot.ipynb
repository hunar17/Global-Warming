```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.offline as py
```

```python
readData= pd.read_csv('GlobalLandTemperatures_GlobalLandTemperaturesByState.csv')
Data1= pd.DataFrame(readData, columns=['dt','AverageTemperature','State','Country'])
Data1
```

```python
Data1= Data1.drop(Data1[(Data1.Country != 'India') & (Data1.Country !='Brazil') & (Data1.Country !='United States')].index)
Data1
```

```python
Data1['dt']= pd.to_datetime(Data1['dt'])
start_date='1900-01-01'
Data1=Data1.loc[(Data1['dt'] > start_date)]
Data1
```

```python
Data1=Data1.dropna()
```

```python
Data1.isnull().sum()
```

```python
Brzil= Data1.loc[Data1['Country'] == 'Brazil']
India= Data1.loc[Data1['Country'] == 'India']
US= Data1.loc[Data1['Country'] == 'United States']

years = np.unique(Brzil['dt'].astype(str).apply(lambda x: x[:4]))
mean_temp_Brazil = []
mean_temp_India = []
mean_temp_US = []

for year in years:
    mean_temp_Brazil.append(Brzil[Brzil['dt'].astype(str).apply(
        lambda x: x[:4]) == year]['AverageTemperature'].mean())
    mean_temp_India.append(India[India['dt'].astype(str).apply(
        lambda x: x[:4]) == year]['AverageTemperature'].mean())
    mean_temp_US.append(US[US['dt'].astype(str).apply(
        lambda x: x[:4]) == year]['AverageTemperature'].mean())

trace0 = go.Scatter(
    x = years, 
    y = mean_temp_US,
    fill= None,
    mode='lines',
    name='US Average Temperature',
    line=dict(
        color='rgb(200, 200, 200)',
    )
)

trace1 = go.Scatter(
    x = years, 
    y = mean_temp_India,
    fill= None,
    mode='lines',
    name='India Average Temperature',
    line=dict(
        color='rgb(100, 100, 100)',
    )
)
trace2 = go.Scatter(
    x = years, 
    y = mean_temp_Brazil,
    name='Brazil Average Temperature',
    line=dict(
        color='rgb(500, 121, 093)',
    )
)
data = [trace0,trace1,trace2]

layout = go.Layout(
    xaxis=dict(title='year'),
    yaxis=dict(title='Average Temperature, °C'),
    title='Average land temperature USA, India & Brazil',
    showlegend = True)

fig = go.Figure(data=data, layout=layout)

py.iplot(fig)
```
