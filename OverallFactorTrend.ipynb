```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
 
 
readData= pd.read_csv('result.csv')
Data1= pd.DataFrame(readData)
dat = Data1.groupby('Country', as_index = False).mean()
dat
 
 
 
Brzil= dat.loc[dat['Country'] == 'Brazil']
Indi= dat.loc[dat['Country'] == 'India']
US= dat.loc[dat['Country'] == 'United States']
 
 
 
labels = ['Brazil', 'India', 'USA']      #, 'Deforestation']
 
 
 
CO2_means =np.array([Brzil['CO2Emissions '], Indi['CO2Emissions '], US['CO2Emissions ']]).ravel()
Population_means =np.array([Brzil['Population (People)'], Indi['Population (People)'], US['Population (People)']]).ravel()
Deforestation_mean = np.array([Brzil['Deforestration'], Indi['Deforestration'], US['Deforestration']]).ravel()
 
 
x = np.arange(len(labels))  # the label locations
width = 0.2  # the width of the bars
 
 
 
fig, ax = plt.subplots()
rects1 = ax.bar(x - 3*width/2, CO2_means, width, log = True, label='CO2')
rects2 = ax.bar(x + width/2, Population_means, width, log = True,label='Population')
react3= ax.bar(x - width/2, Deforestation_mean, width, log = True, label='Deforestation')
 
# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Contribution Trend')
ax.set_title('Contribution of major factors to global warming')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend(bbox_to_anchor=(1.1, 1.05))

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = round(rect.get_height())
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(rects1)
autolabel(rects2)
autolabel(react3)
 
fig.tight_layout()

 
 
 
plt.show()
```
